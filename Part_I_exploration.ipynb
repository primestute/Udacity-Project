### Univariate Exploration
<a id="uni"><a/>

```python
# create a wrange function
def wrangle(filepath):
    # Read csv file
    df = pd.read_csv(filepath)

    # remove outliers
    low, high = df["StatedMonthlyIncome"].quantile([0.1, 0.9])
    mask = df["StatedMonthlyIncome"].between(low, high)
    df = df[mask]
    
    low, high = df["DebtToIncomeRatio"].quantile([0.1, 0.9])
    mask_d = df["DebtToIncomeRatio"].between(low, high)
    df = df[mask_d]
     # drop NaNs Values
    df = df[ (df["Occupation"] != "Other") & (df["Occupation"] != "Professional")\
             & (~df["Occupation"].isna())]
    df = df[(~df.BorrowerAPR.isna())]
    df = df[(~df.IncomeRange.isna())]
    
    return df
```

```python
# use the wrangle function to load the data into dataframe
df = wrangle("prosperLoanData.csv")
```

```python
# Convert ProsperRating and credit grade into ordered categorical types
rate_order = ['HR','E','D','C','B','A','AA']
df['ProsperRating (Alpha)'] = df['ProsperRating (Alpha)'].astype("category")
df['ProsperRating (Alpha)'] = df['ProsperRating (Alpha)'].cat.set_categories(
                                    new_categories = rate_order, ordered = True)

grade_order = ['HR','E','D','C','B','A','AA']
df['CreditGrade'] = df['CreditGrade'].astype("category")
df['CreditGrade'] = df['CreditGrade'].cat.set_categories(
                                    new_categories = rate_order, ordered = True)
```

```python
# create a function to label axes and set tit
def label(xaxis, yaxis, title):
    """This function set the label axes and title of plots"""
    plt.xlabel(xaxis)
    plt.ylabel(yaxis)
    plt.title(title)
    
```

```python
# Summarry statistics
df[["BorrowerAPR", "BorrowerRate"]].describe()
```

```python
# set the figure set
plt.rcParams['figure.figsize'] = 14, 5

fig, ax = plt.subplots(ncols = 2)

bins = np.arange(0, df["BorrowerAPR"].max()+0.01, 0.01)
bin_s = bins = np.arange(0, df["BorrowerRate"].max()+0.01, 0.01)
# plot
ax[0].hist(data = df, x="BorrowerAPR", bins = bins)
ax[1].hist(data = df, x="BorrowerRate", bins = bin_s)

# label the axes
ax[0].set_ylabel("Count")
ax[0].set_xlabel('Interes Rate (%)')
ax[1].set_xlabel("Interest Rate (%)")
ax[1].set_ylabel("Count")

# title
ax[0].set_title("Annual Cost with other Charges")
ax[1].set_title("Annual Cost")

plt.tight_layout()
# show the plot
plt.show()
```

> * Both BorrowerAPR and BorrowerRate interes rate are normally distributed. However, both have spikes.
